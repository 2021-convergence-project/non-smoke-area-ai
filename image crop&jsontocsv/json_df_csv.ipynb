```python
########################## 어그멘테이션 추가 ############################


import pandas as pd
import json
import os
import numpy as np

# 2차원 배열 더하기코드
def sumMatrix(A,B):
    for i in range(len(A)) :
        for j in range(len(A[0])):
            A[i][j] += B[i][j] 
    return A

# 5.csv16행 * 3열 1차원 리스트가 30개 전부 1
data = []

for j in range(40):
    line = []
    for k in range(48):
        line.append(1)
    data.append(line)

path = './team6/2D/5-1'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.json')]

for i in file_list_py:
    with open(path + '/' + i,'r') as j:
        contents=json.loads(j.read())
    # contents = list(contents)

    ID = []
    Keypoints = []
    Person_No = []

    new_Keypoints = []

    for id in contents['images']:
#         print(id['img_path'])
        ID.append(id['img_path'].split("/")[-1][:-8])

    for val in contents['annotations']:
        Person_No.append(val['person_no'])
        Keypoints.append(val['keypoints'])
    
    new_Keypoints = sumMatrix(Keypoints,data)


    df = pd.DataFrame({"ID" : ID*2,
                       "Person_No" : Person_No,
                       "Keypoints" : new_Keypoints})
    
    df.to_csv('5-1.csv',  mode='a', header = False)


# print(Keypoints)


# print(ID)
# print(Keypoints)
# print(new_Keypoints)
```
