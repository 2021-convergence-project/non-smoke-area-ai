## 학습에서 사용하지 않거나 실제 예측에서 사용할 수 없는 keypoint 제거 함수

```python
def remove_aihub_annotation(keypoint):
    # 학습데이터는 모두 2D데이터
    # 따라서, 사용하지 않는 값 (_, _, 2)을 제거
    keypoint = keypoint.reshape(-1, 3)
    result = np.delete(keypoint, 2, axis=1)
    
    # 실제 예측에서 사용할 수 없는 Annotation 제거
    result = np.delete(result, [6,7,8], axis=0)
    result = result.reshape(1,-1)
    
    return result
```

## Dataframe으로 변환한 데이터의 최소/최대/최빈값 출력 함수

```python
def print_mod_min_max(dataframe):
    min_data = dataframe.min()
    max_data = dataframe.max()
    mod_data = dataframe.mode()
    print(max(mod_data.iloc[:, 0]))
    print("/ Min : {} \n/ Max: {} \n/ Mode Max: {}\n".format(min_data, max_data, max(mod_data.iloc[:, 0])))
```

## ID와 Person 데이터를 이용해 각 Person number 별 데이터를 분리 정렬하는 코드
- 최종 결과는 모든 데이터를 취합한 하나의 Dataframe

```python
def seperate_people(dataframe):
    # ['Time', 'ID', 'Person', 'Keypoints']
    max_person_no = dataframe['Person'].max()
    
    if max_person_no==1 :
        return dataframe
    else:
        new_df = pd.DataFrame( columns={'Time', 'ID', 'Person', 'Keypoints'})
        id_data = dataframe[['ID']].drop_duplicates(['ID'])
        id_data = id_data.reset_index(drop=True)

        for k in tqdm(range(0, len(id_data))):
            pre_data = dataframe[ dataframe['ID']==id_data.at[k, 'ID'] ]
            
            for i in range(0, max_person_no):
                data = pre_data[ pre_data['Person']==i+1 ]
                data = data.drop('Time', axis=1)
                data = data.reset_index(drop=True)
                data = data.reset_index()
                data.rename(columns={'index':'Time'}, inplace=True)
#                 print(data)
                new_df = new_df.append(data)
                
        new_df = new_df[['Time', 'ID', 'Person', 'Keypoints']]
        new_df.reset_index(drop=True, inplace=True)
        return new_df
```

## frame 사이즈만큼 남도록 남는 annotation을 삭제하는 코드
- 전체 Time수를 frame으로 나눠 몫을 반올림한 후, 그 값을 이용해 Time을 modular 연산한 결과가 0인 데이터를 남긴다
- 이렇게 하면 데이터가 frame 수만큼 남거나, frame+1 만큼 남게 된다
- 전체 길이를 마지막으로 체크하여 frame보다 클 경우 해당 데이터를 버리고 reindexing 한다

```python
def drop_keypoints_to_size(dataframe, frame):
    # ['Time', 'ID', 'Person', 'Keypoints']
#     max_person_no = dataframe['Person'].max()

    result = pd.DataFrame( columns={'Time', 'ID', 'Person', 'Keypoints'} )

    id_data = dataframe[['ID']].drop_duplicates(['ID'])
    id_data = id_data.reset_index(drop=True)

#     for k in tqdm(range(0, 3)):
    for k in tqdm(range(0, len(id_data))):
        pre_data = dataframe[ dataframe['ID']==id_data.at[k, 'ID'] ]
        max_person_no = pre_data['Person'].max()

        for i in range(0, max_person_no):
            data = pre_data[ pre_data['Person']==i+1 ]
    #         print(data)
            max_time_no = data['Time'].max() +1
            div_size = (max_time_no / frame)

            if div_size<1:
                # 주어진 사이즈보다 작음
                pass
            else:
                div_size = round(div_size)
#                 print(div_size)

                # 필요 이상의 값을 찾아와 drop
                drop_index = data[ data['Time']%div_size>0 ].index
                data.drop(drop_index, inplace=True)

                # Time 값 data
                data = data.drop('Time', axis=1)
                data = data.reset_index(drop=True)
                data = data.reset_index()
                data.rename(columns={'index':'Time'}, inplace=True)
#             print(data[['Time', 'ID', 'Person']])
#             print("len(data)")
#             print(len(data))
            data_len = len(data)

            if data_len>=(frame+1) :
                # 지정 개수 이상일 시 남는 데이터 drop
                drop_index = data[ data['Time']>=frame ].index
#                 print(drop_index)
                data.drop(drop_index, inplace=True)
#                 data.drop(frame, inplace=True)
            
            elif data_len<(frame) :
                # 지정 개수 미만일 시 남는 마지막 데이터를 복사해서 붙인다
                for i in range(0, frame-data_len):
                    data = data.append(data.loc[ data_len-1 ])
                
                data = data.drop('Time', axis=1)
                data = data.reset_index(drop=True)
                data = data.reset_index()
                data.rename(columns={'index':'Time'}, inplace=True)
#             print(data)
#             print("len(data) (2)")
#             print(len(data))
            result = result.append(data)

    result = result[['Time', 'ID', 'Person', 'Keypoints']]
    result.reset_index(drop=True, inplace=True)
    return result


# test_index = df2[df2['ID'].str.startswith('45-')].index
# test_df = df2.loc[test_index]
# test_df.reset_index(drop=True, inplace=True)

# new_test_df = drop_keypoints_to_size(test_df, 8)
# new_test_df


# print(len(new_df_5))
# new_df_5_dropped = drop_keypoints_to_size(new_df_5, 8)
# new_df_5_dropped.head(50)

# new_df_5_dropped.describe()
# result.head(50)

# new_df['Time'].describe()
# new_df[ new_df['Time']==8 ]
```

## ID 데이터를 이용해 train/test set을 분리하는 코드
- split_my_train_test(dataframe):
    - 인덱스 형식 = `['Time', 'ID', 'Person', 'Keypoints', 'Answer']`
- split_my_train_test_set(dataframe):
    - 인덱스 형식 = `['Time', 'ID', 'Keypoints', 'Answer']`

```python
def split_my_train_test(dataframe):
    # ['Time', 'ID', 'Person', 'Keypoints']

    id_data = dataframe[['ID', 'Answer']].drop_duplicates(['ID'])
    id_data = id_data.reset_index(drop=True)

    train_x_data, test_x_data, train_t_data, test_t_data = \
    train_test_split(id_data.drop('Answer', axis=1, inplace=False),
                     id_data['Answer'],
                     test_size=0.3,
                     stratify=id_data['Answer'],
                     random_state=0)
    train_x_data.reset_index(drop=True, inplace=True)
    test_x_data.reset_index(drop=True, inplace=True)
    
    result_train = pd.DataFrame( columns={'Time', 'ID', 'Person', 'Keypoints', 'Answer'} )
    result_test = pd.DataFrame( columns={'Time', 'ID', 'Person', 'Keypoints', 'Answer'} )

#     for k in tqdm(range(0, 3)):
    for k in tqdm(range(0, len(train_x_data))):
        pre_data = dataframe[ dataframe['ID']==train_x_data.at[k, 'ID'] ]
        result_train = result_train.append(pre_data)
    
    for k in tqdm(range(0, len(test_x_data))):
        pre_data = dataframe[ dataframe['ID']==test_x_data.at[k, 'ID'] ]
        result_test = result_test.append(pre_data)

    result_train = result_train[['Time', 'ID', 'Person', 'Keypoints', 'Answer']]
    result_test = result_test[['Time', 'ID', 'Person', 'Keypoints', 'Answer']]
    
    result_train.reset_index(drop=True, inplace=True)
    result_test.reset_index(drop=True, inplace=True)
    
#     result_train.to_csv('./train/dropped_train.csv', mode='w', header = True, index=False)
#     result_test.to_csv('./train/dropped_test.csv', mode='w', header = True, index=False)
    return result_train, result_test, train_t_data, test_t_data
```

```python
def split_my_train_test_set(dataframe):
    # ['Time', 'ID', 'Keypoints', 'Answer']

    id_data = dataframe[['ID', 'Answer']].drop_duplicates(['ID'])
    id_data = id_data.reset_index(drop=True)

    train_x_data, test_x_data, train_t_data, test_t_data = \
    train_test_split(id_data.drop('Answer', axis=1, inplace=False),
                     id_data['Answer'],
                     test_size=0.3,
                     stratify=id_data['Answer'],
                     random_state=0)
    train_x_data.reset_index(drop=True, inplace=True)
    test_x_data.reset_index(drop=True, inplace=True)
    train_t_data.reset_index(drop=True, inplace=True)
    test_t_data.reset_index(drop=True, inplace=True)
    
    result_train = pd.DataFrame( columns={'Time', 'ID', 'Keypoints', 'Answer'} )
    result_test = pd.DataFrame( columns={'Time', 'ID', 'Keypoints', 'Answer'} )

#     for k in tqdm(range(0, 3)):
    for k in tqdm(range(0, len(train_x_data))):
        pre_data = dataframe[ dataframe['ID']==train_x_data.at[k, 'ID'] ]
        result_train = result_train.append(pre_data)
    
    for k in tqdm(range(0, len(test_x_data))):
        pre_data = dataframe[ dataframe['ID']==test_x_data.at[k, 'ID'] ]
        result_test = result_test.append(pre_data)

    result_train = result_train[['Time', 'ID', 'Keypoints', 'Answer']]
    result_test = result_test[['Time', 'ID', 'Keypoints', 'Answer']]
    
    result_train.reset_index(drop=True, inplace=True)
    result_test.reset_index(drop=True, inplace=True)
    
    return result_train, result_test, train_t_data, test_t_data
```

## Dataframe에서 가져온 str 형태의 데이터를 float array로 변환하는 함수

```python
# KEYPOINT = 32
# KEYPOINT_MOD = 26

def get_keypoints_in_float(string_data):
    keypoint = np.array(string_data['Keypoints'].values)
    print(keypoint.shape)
    keypoint_arr = np.empty((0,KEYPOINT_MOD), float)

#     i=0
    for i in tqdm(range(0, len(keypoint))) :
        keypoint[i] = keypoint[i].replace("[", "")
        keypoint[i] = keypoint[i].replace("]", "")
        keypoint[i] = keypoint[i].split(',')

        # keypoint[0]
        nd_array = np.array(keypoint[i])
        nd_array = nd_array.astype(np.float)
    #     print("nd_array")
    #     print(nd_array)
    #     print(nd_array.shape)

        # 학습데이터는 모두 2D데이터
        # 사용하지 않는 값과 사용할 수 없는 annotation을 제거
        result = remove_aihub_annotation(nd_array)
    #     print(result)
    #     print(result.shape)
    #     print(type(result[0][0]))

        keypoint_arr = np.append(keypoint_arr, result, axis=0)
    return keypoint_arr
```
